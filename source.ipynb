```python
#Removing column with one unique value which will have no impaact on model

for column in data.columns:
    if(data[column].nunique() == 1) :
        data.drop(column,axis="columns", inplace=True)
```

```python
#dropping time variable as we are not processing time dependent task
data.drop(["time","source_name"],axis="columns", inplace=True)
```

```python
target = data['y_true(fc)']
data.drop('y_true(fc)',axis="columns", inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
# Converting String dtype to float64 dtype
le = LabelEncoder()

target = le.fit_transform(target)
```

```python
import numpy as np

for column in data.columns:
    # Ensure that the columns you want to process are numeric (e.g., float or int)
    #data[column] = data[column].astype(float)

    upper_quartile = np.percentile(data[column], 75)
    lower_quartile = np.percentile(data[column], 25)
    IQR = (upper_quartile - lower_quartile) * 1.5
    quartileSet = (lower_quartile - IQR, upper_quartile + IQR)

    data[column] = np.where(
      data[column]< quartileSet[0],
      0,
      np.where(
          data[column]>quartileSet[1],
          0,
          data[column]
    )
    )
data = data[(data != 0).any(axis=1)]
```
